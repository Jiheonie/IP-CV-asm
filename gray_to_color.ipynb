```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pic_r = Image.open('puppy_red.jpg')
pic_g = Image.open('puppy_green.jpg')
pic_b = Image.open('puppy_blue.jpg')
```

```python
r_arr= np.asarray(pic_r)
g_arr= np.asarray(pic_g)
b_arr= np.asarray(pic_b)
```

```python
r_arr.shape
```

```python
g_arr.shape
```

```python
b_arr.shape
```

```python
new_shape = (r_arr.shape[0], r_arr.shape[1], 3)
img_arr = np.zeros(new_shape, dtype=int)
```

```python
for i in range(new_shape[0]):
  for j in range(new_shape[1]):
    img_arr[i][j][0] = r_arr[i][j]
    img_arr[i][j][1] = g_arr[i][j]    
    img_arr[i][j][2] = b_arr[i][j]  
```
